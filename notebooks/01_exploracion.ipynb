## 3. Analisis de Nulos:

````python
# Obtencion de nulos por columna.

print(f"{'='*40} NULOS POR COLUMNA {'='*40}")

cant_nulos = df.isna().sum()
pct_nulos = (cant_nulos / len(df)*100).round(4)

reporte_nulos = pd.DataFrame({
    'nulos': cant_nulos,
    'porcentaje_(%)': pct_nulos
})

display(reporte_nulos[reporte_nulos['nulos'] > 0])
````

Hallazgo:

+ La columna `MUNICIPIO` presenta `4 valores nulos` (0.0006%).

+ La columna `GENERO` presenta `2606 valores nulos` (0.3944%).

+ La columna `GRUPO ETARIO` presenta `1024 valores nulos` (0.1550%).

En todos los casos, la proporción de valores nulos es `< 1%`, lo que indica una baja incidencia de datos faltantes en el dataset.

## 4. Duplicados Exactos:

````python

print(f"{'='*40} DUPLICADOS EXACTOS {'='*40}")

cant_duplicados = df.duplicated().sum()
pct_duplicados = (cant_duplicados / len(df)*100).round(2)

print(f"\nLa cantidad de duplicados exactos es {cant_duplicados}, lo cual representa el {pct_duplicados}% de todo el Dataset.")
````

**Hallazgo:**

El dataset en estado crudo no presenta duplicados completos (registros identicos en todas sus columnas).

Sin embargo, esta validación debe repetirse en el cuaderno `02_limpieza.ipynb` despues de aplicar los procesos de normalizacion y estandarizacion, dado que estas transformaciones pueden revelar duplicados no evidentes en la etapa inicial (por ejemplo, diferencias en mayusculas/minusculas, tildes, formatos o espacios).

## 7. Rango Temporal.

Se verifica que las fechas correspondan al periodo declarado del dataset (01/01/2010 - 31/10/2025).

````python

print(f"Fecha de inicio: {pd.to_datetime(df['FECHA HECHO'], format="%d/%m/%Y", dayfirst=True, errors='coerce').min()}")
print(f"Fecha de fin: {pd.to_datetime(df['FECHA HECHO'], format="%d/%m/%Y", dayfirst=True, errors='coerce').max()}")

````

**Hallazgo:**

La columna `FECHA HECHO` se encuentra en formato texto, lo que genera una evaluacion incorrecta del rango de fechas.

* **Evidencia:**  
Al aplicar la funcion `max()` sin parseo previo, la fecha maxima obtenida fue `31/12/2024`, debido a comparacion lexicografica y no cronologica.

* **Causa:**  
Las fechas no estan en tipo `datetime`, sino en tipo `object` (string).

* **Accion:**  
Se realizo el parseo de la columna a tipo `datetime` unicamente para efectos de validacion del rango de fechas, sin persistir la transformacion en el dataset.

* **Resultado:**  
Posterior al parseo, la fecha maxima corresponde al valor real y el rango de fechas es consistente.
